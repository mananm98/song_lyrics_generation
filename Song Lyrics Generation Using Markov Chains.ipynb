## creating dictionary of probabilities

```python
def create_dictionary(text,k=4):
    T = {}
    
    for i in range(len(text) - k):
        X = text[i:i+k]
        Y = text[i+k]
        
        if T.get(X) is None:
            T[X] = {}
            T[X][Y] = 1
           
        else:
            if T[X].get(Y) is None:
                T[X][Y] = 1
               
            else:
                T[X][Y] += 1
               
           
    # finding probabilities of each occurence
    for key in T.keys():
        s = sum([val for val in T[key].values()])
        for kx in T[key].keys():
            T[key][kx] = T[key][kx]/s
           
       
    return T

            
    
            
```

## Sampling 

```python
def sample_next(text,model,k=4):
    
    if model.get(text) is None:
        return ' '
        
    possible_chars = list(model[text].keys())
    possible_vals = list(model[text].values())
    
    return np.random.choice(possible_chars,p = possible_vals)
```

## Generating new text

```python
def generate_newtext(sentence,model,maxlen,k = 4):
    text = sentence[-k:]
    
    for i in range(maxlen):
        next_char = sample_next(text,model)
        sentence = sentence + next_char
        text = sentence[-k:]
    
    return sentence
```
